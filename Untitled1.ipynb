```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

class JLawTradingSystem:
    def __init__(self, symbol, start, end, ma_short=50, ma_long=200, 
                 stop_loss=0.08, profit_target=0.20, position_size=0.20):
        self.symbol = symbol
        self.start = start
        self.end = end
        self.ma_short = ma_short
        self.ma_long = ma_long
        self.stop_loss = stop_loss
        self.profit_target = profit_target
        self.position_size = position_size
        self.results = None
        self.trades = []
        self.get_data()
        
    def get_data(self):
        df = yf.download(self.symbol, start=self.start, end=self.end)
        data = df[['Open', 'High', 'Low', 'Close', 'Volume']].copy()
        
        data['returns'] = np.log(data['Close'] / data['Close'].shift(1))
        
        data['MA50'] = data['Close'].rolling(window=self.ma_short).mean()
        data['MA200'] = data['Close'].rolling(window=self.ma_long).mean()
        
        data['stage'] = self.identify_stage(data)
        
        data['volume_ma'] = data['Volume'].rolling(window=20).mean()
        data['volume_spike'] = data['Volume'] > (data['volume_ma'] * 1.5)
        
        data['daily_range'] = (data['High'] - data['Low']) / data['Close']
        data['body_size'] = abs(data['Close'] - data['Open']) / data['Close']
        data['upper_shadow'] = (data['High'] - np.maximum(data['Open'], data['Close'])) / data['Close']
        data['lower_shadow'] = (np.minimum(data['Open'], data['Close']) - data['Low']) / data['Close']
        
        data['momentum_5'] = data['Close'] / data['Close'].shift(5) - 1
        data['momentum_10'] = data['Close'] / data['Close'].shift(10) - 1
        
        data['resistance'] = data['High'].rolling(window=20).max()
        data['support'] = data['Low'].rolling(window=20).min()
        
        try:
            spy = yf.download('SPY', start=self.start, end=self.end)['Close']
            spy_returns = spy.pct_change()
            stock_returns = data['Close'].pct_change()
            
            # 確保日期對齊
            aligned_spy_returns = spy_returns.reindex(data.index).fillna(0)
            data['relative_strength'] = (stock_returns - aligned_spy_returns).rolling(window=20).mean()
        except:
            data['relative_strength'] = 0
            
        data.dropna(inplace=True)
        self.data = data
        return data
    
    def identify_stage(self, data):
        stages = pd.Series(index=data.index, dtype=int)
        stages[:] = 1  # 初始化所有值為1
        
        for i in range(len(data)):
            if i < self.ma_long:
                stages.iloc[i] = 1
                continue
                
            price = data['Close'].iloc[i]
            ma50 = data['MA50'].iloc[i] if i >= self.ma_short and not pd.isna(data['MA50'].iloc[i]) else price
            ma200 = data['MA200'].iloc[i]
            
            # 檢查是否有NaN值
            if pd.isna(ma200) or pd.isna(ma50):
                stages.iloc[i] = 1
                continue
            
            # 檢查MA200趨勢（確保有足夠的歷史數據）
            ma200_trend_up = False
            ma200_trend_down = False
            
            if i >= self.ma_long + 10:
                ma200_trend_up = data['MA200'].iloc[i] > data['MA200'].iloc[i-10]
                ma200_trend_down = data['MA200'].iloc[i] < data['MA200'].iloc[i-10]
            
            if (price > ma200 and ma50 > ma200 and ma200_trend_up):
                stages.iloc[i] = 2
            elif (price < ma200 and ma50 < ma200 and ma200_trend_down):
                stages.iloc[i] = 4
            elif price > ma200 and ma50 < ma200:
                stages.iloc[i] = 3
            else:
                stages.iloc[i] = 1
                
        return stages
    
    def identify_multiple_edges(self, i):
        edges = 0
        edge_details = []
        
        if i < 20:
            return edges, edge_details
            
        data = self.data
        
        # 檢查所有需要的值是否存在且不為NaN
        required_values = [
            data['stage'].iloc[i],
            data['momentum_5'].iloc[i],
            data['volume_spike'].iloc[i],
            data['Close'].iloc[i],
            data['Open'].iloc[i],
            data['body_size'].iloc[i],
            data['resistance'].iloc[i-1],
            data['relative_strength'].iloc[i],
            data['MA50'].iloc[i]
        ]
        
        if any(pd.isna(val) for val in required_values if not isinstance(val, bool)):
            return edges, edge_details
        
        if (data['stage'].iloc[i] == 2 and 
            data['momentum_5'].iloc[i] > 0.02):
            edges += 1
            edge_details.append("Strong uptrend momentum")
            
        if data['volume_spike'].iloc[i]:
            edges += 1
            edge_details.append("Volume breakout")
            
        if (data['Close'].iloc[i] > data['Open'].iloc[i] and 
            data['body_size'].iloc[i] > 0.02):
            edges += 1
            edge_details.append("Strong bullish candle")
            
        if (data['Close'].iloc[i] > data['resistance'].iloc[i-1] and 
            data['Close'].iloc[i-1] <= data['resistance'].iloc[i-1]):
            edges += 1
            edge_details.append("Resistance breakout")
            
        if data['relative_strength'].iloc[i] > 0.01:
            edges += 1
            edge_details.append("Relative strength advantage")
            
        if (data['Close'].iloc[i] > data['MA50'].iloc[i] and 
            data['Close'].iloc[i-1] <= data['MA50'].iloc[i-1]):
            edges += 1
            edge_details.append("MA50 support reclaim")
            
        return edges, edge_details
    
    def execute_trading_strategy(self):
        data = self.data.copy()
        data['position'] = 0.0
        data['entry_price'] = 0.0
        data['stop_loss_price'] = 0.0
        data['edges_count'] = 0
        
        current_position = 0.0
        entry_price = 0.0
        stop_loss_price = 0.0
        
        for i in range(20, len(data)):
            current_price = data['Close'].iloc[i]
            
            if current_position > 0:
                if current_price <= stop_loss_price:
                    self.record_trade(entry_price, current_price, current_position, 'Stop Loss', i)
                    current_position = 0.0
                    entry_price = 0.0
                    stop_loss_price = 0.0
                    
                elif current_price >= entry_price * (1 + self.profit_target):
                    exit_size = current_position * 0.5
                    self.record_trade(entry_price, current_price, exit_size, 'Profit Target', i)
                    current_position -= exit_size
                    stop_loss_price = entry_price
                    
                elif (data['stage'].iloc[i] in [3, 4] and 
                      current_price < data['MA50'].iloc[i]):
                    self.record_trade(entry_price, current_price, current_position, 'Trend Change', i)
                    current_position = 0.0
                    entry_price = 0.0
                    stop_loss_price = 0.0
            
            if current_position == 0:
                edges_count, edge_details = self.identify_multiple_edges(i)
                
                if edges_count >= 3 and data['stage'].iloc[i] == 2:
                    current_position = self.position_size
                    entry_price = current_price
                    stop_loss_price = entry_price * (1 - self.stop_loss)
                    
                data.iloc[i, data.columns.get_loc('edges_count')] = edges_count
            
            data.iloc[i, data.columns.get_loc('position')] = current_position
            data.iloc[i, data.columns.get_loc('entry_price')] = entry_price
            data.iloc[i, data.columns.get_loc('stop_loss_price')] = stop_loss_price
        
        data['strategy_returns'] = data['returns'] * data['position'].shift(1)
        data['cum_returns'] = data['returns'].cumsum()
        data['cum_strategy_returns'] = data['strategy_returns'].cumsum()
        data['total_returns'] = np.exp(data['cum_returns'])
        data['total_strategy_returns'] = np.exp(data['cum_strategy_returns'])
        
        self.results = data
        return data
    
    def record_trade(self, entry_price, exit_price, position_size, exit_reason, exit_index):
        pnl = (exit_price - entry_price) / entry_price * position_size
        self.trades.append({
            'exit_date': self.data.index[exit_index],
            'exit_price': exit_price,
            'position_size': position_size,
            'pnl': pnl,
            'exit_reason': exit_reason
        })
    
    def analyze_performance(self):
        if self.results is None:
            print("Please run the strategy first")
            return
            
        data = self.results
        
        total_return = data['total_strategy_returns'].iloc[-1] - 1
        buy_hold_return = data['total_returns'].iloc[-1] - 1
        outperformance = total_return - buy_hold_return
        
        strategy_volatility = data['strategy_returns'].std() * np.sqrt(252)
        max_drawdown = self.calculate_max_drawdown(data['total_strategy_returns'])
        
        winning_trades = [t for t in self.trades if t['pnl'] > 0]
        losing_trades = [t for t in self.trades if t['pnl'] < 0]
        
        win_rate = len(winning_trades) / len(self.trades) if self.trades else 0
        avg_win = np.mean([t['pnl'] for t in winning_trades]) if winning_trades else 0
        avg_loss = np.mean([t['pnl'] for t in losing_trades]) if losing_trades else 0
        profit_factor = abs(avg_win / avg_loss) if avg_loss != 0 else 0
        
        print(f"\n=== J Law Trading System Performance Analysis ===")
        print(f"Symbol: {self.symbol}")
        print(f"Period: {self.start} to {self.end}")
        print(f"\nStrategy Performance:")
        print(f"Total Return: {total_return:.2%}")
        print(f"Buy & Hold Return: {buy_hold_return:.2%}")
        print(f"Outperformance: {outperformance:.2%}")
        print(f"Volatility: {strategy_volatility:.2%}")
        print(f"Max Drawdown: {max_drawdown:.2%}")
        print(f"\nTrade Analysis:")
        print(f"Total Trades: {len(self.trades)}")
        print(f"Win Rate: {win_rate:.2%}")
        print(f"Average Win: {avg_win:.2%}")
        print(f"Average Loss: {avg_loss:.2%}")
        print(f"Profit Factor: {profit_factor:.2f}")
        
        return {
            'total_return': total_return,
            'buy_hold_return': buy_hold_return,
            'outperformance': outperformance,
            'volatility': strategy_volatility,
            'max_drawdown': max_drawdown,
            'total_trades': len(self.trades),
            'win_rate': win_rate,
            'profit_factor': profit_factor
        }
    
    def calculate_max_drawdown(self, returns_series):
        peak = returns_series.expanding().max()
        drawdown = (returns_series - peak) / peak
        return drawdown.min()
    
    def plot_results(self):
        if self.results is None:
            print("Please run the strategy first")
            return
            
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
        
        ax1.plot(self.results.index, self.results['Close'], label='Close Price', linewidth=1)
        ax1.plot(self.results.index, self.results['MA50'], label='MA50', alpha=0.7)
        ax1.plot(self.results.index, self.results['MA200'], label='MA200', alpha=0.7)
        
        entries = self.results[self.results['position'] > self.results['position'].shift(1).fillna(0)]
        if not entries.empty:
            ax1.scatter(entries.index, entries['Close'], color='green', marker='^', s=50, label='Entry')
        
        ax1.set_title(f'{self.symbol} - Price Action with Moving Averages')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        ax2.plot(self.results.index, self.results['total_returns'], label='Buy & Hold', linewidth=2)
        ax2.plot(self.results.index, self.results['total_strategy_returns'], label='J Law Strategy', linewidth=2)
        ax2.set_title('Cumulative Returns Comparison')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        ax3.plot(self.results.index, self.results['position'], label='Position Size', color='orange')
        ax3.set_title('Position Sizing Over Time')
        ax3.set_ylabel('Position Size')
        ax3.legend()
        ax3.grid(True, alpha=0.3)
        
        ax4.plot(self.results.index, self.results['edges_count'], label='Edges Count', color='red', alpha=0.7)
        ax4.axhline(y=3, color='black', linestyle='--', alpha=0.5, label='Entry Threshold')
        ax4.set_title('Multiple Trading Edges Count')
        ax4.set_ylabel('Number of Edges')
        ax4.legend()
        ax4.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    def plot_stage_analysis(self):
        if self.results is None:
            print("Please run the strategy first")
            return
            
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10))
        
        stage_colors = {1: 'gray', 2: 'green', 3: 'orange', 4: 'red'}
        
        for stage in [1, 2, 3, 4]:
            stage_data = self.results[self.results['stage'] == stage]
            if not stage_data.empty:
                ax1.scatter(stage_data.index, stage_data['Close'], 
                           c=stage_colors[stage], alpha=0.6, s=10, 
                           label=f'Stage {stage}')
        
        ax1.plot(self.results.index, self.results['Close'], color='black', alpha=0.3, linewidth=0.5)
        ax1.set_title(f'{self.symbol} - Market Stages Analysis')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        stage_counts = self.results['stage'].value_counts().sort_index()
        ax2.bar(stage_counts.index, stage_counts.values, color=[stage_colors[i] for i in stage_counts.index])
        ax2.set_title('Market Stage Distribution')
        ax2.set_xlabel('Stage')
        ax2.set_ylabel('Number of Days')
        ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
```
